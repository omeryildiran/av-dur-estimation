```python
# Statistical Analysis Functions
import scipy.stats as stats
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm

def statistical_analysis_conflict_vs_pse(best_fit, allBootedFits):
    """
    Comprehensive statistical analysis of conflict vs PSE relationship
    """
    print("="*60)
    print("STATISTICAL ANALYSIS: CONFLICT vs PSE")
    print("="*60)
    
    # Collect data for analysis
    analysis_data = []
    
    for i, standardLevel in enumerate(uniqueStandard):
        for j, audioNoiseLevel in enumerate(uniqueSensory):
            conflict_levels = []
            pse_values = []
            pse_ci_lower = []
            pse_ci_upper = []
            
            for k, conflictLevel in enumerate(uniqueConflict):
                lambda_, mu, sigma = getParams(best_fit.x, conflictLevel, audioNoiseLevel, nLambda, nSigma)
                
                # Convert to ms and apply the /2 factor as in the plot
                pse_ms = mu * 1000 / 2
                
                # Calculate confidence intervals from bootstrap
                mu_all = []
                for fit in allBootedFits:
                    lambda_, muBooted, sigma = getParams(fit, conflictLevel, audioNoiseLevel, nLambda, nSigma)
                    mu_all.append(muBooted * 1000 / 2)
                
                mu_ci = np.percentile(mu_all, [2.5, 97.5])
                
                conflict_levels.append(conflictLevel * 1000)  # Convert to ms
                pse_values.append(pse_ms)
                pse_ci_lower.append(mu_ci[0])
                pse_ci_upper.append(mu_ci[1])
                
                analysis_data.append({
                    'standard_level': standardLevel,
                    'noise_level': audioNoiseLevel,
                    'conflict_ms': conflictLevel * 1000,
                    'pse_ms': pse_ms,
                    'pse_ci_lower': mu_ci[0],
                    'pse_ci_upper': mu_ci[1]
                })
            
            # Perform statistical tests for this noise level
            print(f"\nNOISE LEVEL: {audioNoiseLevel}")
            print("-" * 40)
            
            # 1. Linear regression
            X = np.array(conflict_levels).reshape(-1, 1)
            y = np.array(pse_values)
            
            reg = LinearRegression().fit(X, y)
            slope = reg.coef_[0]
            intercept = reg.intercept_
            r_squared = reg.score(X, y)
            
            print(f"Linear Regression:")
            print(f"  Slope: {slope:.4f} ms PSE per ms conflict")
            print(f"  Intercept: {intercept:.4f} ms")
            print(f"  R²: {r_squared:.4f}")
            
            # 2. Correlation analysis
            pearson_r, pearson_p = pearsonr(conflict_levels, pse_values)
            spearman_r, spearman_p = spearmanr(conflict_levels, pse_values)
            
            print(f"\nCorrelation Analysis:")
            print(f"  Pearson r: {pearson_r:.4f}, p = {pearson_p:.4f}")
            print(f"  Spearman ρ: {spearman_r:.4f}, p = {spearman_p:.4f}")
            
            # 3. Test if slope is significantly different from 0
            X_with_const = sm.add_constant(X.flatten())
            model = sm.OLS(y, X_with_const).fit()
            
            print(f"\nSlope Significance Test:")
            print(f"  t-statistic: {model.tvalues[1]:.4f}")
            print(f"  p-value: {model.pvalues[1]:.4f}")
            conf_int = model.conf_int()
            print(f"  95% CI for slope: [{conf_int[1, 0]:.4f}, {conf_int[1, 1]:.4f}]")
            
            # 4. Effect size (Cohen's conventions for correlation)
            if abs(pearson_r) < 0.1:
                effect_size = "negligible"
            elif abs(pearson_r) < 0.3:
                effect_size = "small"
            elif abs(pearson_r) < 0.5:
                effect_size = "medium"
            else:
                effect_size = "large"
            
            print(f"  Effect size: {effect_size}")
    
    # Convert to DataFrame for easier manipulation
    df_analysis = pd.DataFrame(analysis_data)
    
    # Between-noise-level comparison
    print(f"\n" + "="*60)
    print("BETWEEN NOISE LEVELS COMPARISON")
    print("="*60)
    
    # Test if slopes differ between noise levels
    slopes_by_noise = {}
    for noise_level in uniqueSensory:
        subset = df_analysis[df_analysis['noise_level'] == noise_level]
        X = subset['conflict_ms'].values.reshape(-1, 1)
        y = subset['pse_ms'].values
        reg = LinearRegression().fit(X, y)
        slopes_by_noise[noise_level] = reg.coef_[0]
    
    print(f"Slopes by noise level:")
    for noise, slope in slopes_by_noise.items():
        print(f"  {noise}: {slope:.4f} ms PSE per ms conflict")
    
    # Statistical test for difference in slopes
    noise_levels = list(slopes_by_noise.keys())
    if len(noise_levels) == 2:
        slope_diff = slopes_by_noise[noise_levels[0]] - slopes_by_noise[noise_levels[1]]
        print(f"\nSlope difference: {slope_diff:.4f}")
        
        # Perform ANCOVA to test interaction between noise level and conflict
        df_long = df_analysis.copy()
        df_long['noise_categorical'] = df_long['noise_level'].astype(str)
        
        # Fit models for ANCOVA
        model_main = sm.OLS.from_formula('pse_ms ~ conflict_ms + C(noise_categorical)', data=df_long).fit()
        model_interaction = sm.OLS.from_formula('pse_ms ~ conflict_ms * C(noise_categorical)', data=df_long).fit()
        
        # F-test for interaction
        f_stat = ((model_main.ssr - model_interaction.ssr) / 1) / (model_interaction.ssr / model_interaction.df_resid)
        f_p = 1 - stats.f.cdf(f_stat, 1, model_interaction.df_resid)
        
        print(f"\nANCOVA - Interaction Test:")
        print(f"  F-statistic: {f_stat:.4f}")
        print(f"  p-value: {f_p:.4f}")
        
        if f_p < 0.05:
            print("  → Significant interaction: slopes differ between noise levels")
        else:
            print("  → No significant interaction: slopes do not differ significantly")
    
    return df_analysis

def bootstrap_slope_analysis(best_fit, allBootedFits):
    """
    Bootstrap analysis of slope reliability
    """
    print(f"\n" + "="*60)
    print("BOOTSTRAP SLOPE ANALYSIS")
    print("="*60)
    
    slopes_bootstrap = {noise: [] for noise in uniqueSensory}
    
    # Calculate slopes for each bootstrap iteration
    for fit in allBootedFits:
        for noise_level in uniqueSensory:
            conflict_levels = []
            pse_values = []
            
            for conflictLevel in uniqueConflict:
                lambda_, mu, sigma = getParams(fit, conflictLevel, noise_level, nLambda, nSigma)
                conflict_levels.append(conflictLevel * 1000)
                pse_values.append(mu * 1000 / 2)
            
            # Calculate slope
            X = np.array(conflict_levels).reshape(-1, 1)
            y = np.array(pse_values)
            reg = LinearRegression().fit(X, y)
            slopes_bootstrap[noise_level].append(reg.coef_[0])
    
    # Analyze bootstrap slopes
    for noise_level in uniqueSensory:
        slopes = np.array(slopes_bootstrap[noise_level])
        mean_slope = np.mean(slopes)
        std_slope = np.std(slopes)
        ci_lower, ci_upper = np.percentile(slopes, [2.5, 97.5])
        
        print(f"\nNoise Level {noise_level}:")
        print(f"  Bootstrap slope mean: {mean_slope:.4f} ± {std_slope:.4f}")
        print(f"  95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
        
        # Test if slope is significantly different from 0
        t_stat = mean_slope / (std_slope / np.sqrt(len(slopes)))
        p_val = 2 * (1 - stats.t.cdf(abs(t_stat), len(slopes) - 1))
        
        print(f"  t-test vs 0: t = {t_stat:.4f}, p = {p_val:.4f}")
        
        if p_val < 0.001:
            sig_level = "***"
        elif p_val < 0.01:
            sig_level = "**"
        elif p_val < 0.05:
            sig_level = "*"
        else:
            sig_level = "ns"
        
        print(f"  Significance: {sig_level}")
    
    return slopes_bootstrap

def plot_enhanced_conflict_vs_pse(best_fit, allBootedFits):
    """
    Enhanced version with statistical annotations
    """
    plt.figure(figsize=(15, 8))
    
    # Perform statistical analysis first
    df_analysis = statistical_analysis_conflict_vs_pse(best_fit, allBootedFits)
    slopes_bootstrap = bootstrap_slope_analysis(best_fit, allBootedFits)
    
    m = 0
    for i, standardLevel in enumerate(uniqueStandard):
        for j, audioNoiseLevel in enumerate(uniqueSensory):
            conflict_levels = []
            pse_values = []
            colors_list = []
            
            for k, conflictLevel in enumerate(uniqueConflict):
                lambda_, mu, sigma = getParams(best_fit.x, conflictLevel, audioNoiseLevel, nLambda, nSigma)
                m += 1        
                plt.subplot(1, 2, j + 1)
                color = sns.color_palette("viridis", as_cmap=True)(k / len(uniqueConflict))
                if audioNoiseLevel == 0.1:
                    color = sns.color_palette("Reds", as_cmap=True)(0.2 + 0.8 * (k / len(uniqueConflict)))
                elif audioNoiseLevel == 1.2:
                    color = sns.color_palette("Blues", as_cmap=True)(0.2+ 0.8 * (k / len(uniqueConflict)))
                
                conflict_ms = conflictLevel * 1000
                pse_ms = mu * 1000 / 2
                
                conflict_levels.append(conflict_ms)
                pse_values.append(pse_ms)
                colors_list.append(color)
                
                plt.scatter(conflict_ms, pse_ms, color=color, s=100)
                
                # Add error bars
                mu_all = []
                for fit in allBootedFits:
                    lambda_, muBooted, sigma = getParams(fit, conflictLevel, audioNoiseLevel, nLambda, nSigma)
                    mu_all.append(muBooted)
                mu_ci = np.percentile(mu_all, [2.5, 97.5])
                lower_err = np.maximum(pse_ms-mu_ci[0]*1000/2, 0)
                upper_err = np.maximum(mu_ci[1]*1000/2-pse_ms, 0)
                
                plt.errorbar(conflict_ms, pse_ms, yerr=[[lower_err], [upper_err]], 
                            fmt='o', color=color, capsize=10, label=conflictLevel*1000, linewidth=2)
            
            # Add regression line
            if len(conflict_levels) > 1:
                X = np.array(conflict_levels)
                y = np.array(pse_values)
                
                # Fit line
                reg = LinearRegression().fit(X.reshape(-1, 1), y)
                x_line = np.linspace(min(conflict_levels), max(conflict_levels), 100)
                y_line = reg.predict(x_line.reshape(-1, 1))
                
                # Get slope statistics
                subset = df_analysis[df_analysis['noise_level'] == audioNoiseLevel]
                pearson_r, pearson_p = pearsonr(subset['conflict_ms'], subset['pse_ms'])
                
                # Plot regression line
                plt.plot(x_line, y_line, 'k--', alpha=0.7, linewidth=2)
                
                # Add statistical annotation
                if pearson_p < 0.001:
                    sig_text = "***"
                elif pearson_p < 0.01:
                    sig_text = "**"
                elif pearson_p < 0.05:
                    sig_text = "*"
                else:
                    sig_text = "ns"
                
                stats_text = f'r = {pearson_r:.3f}{sig_text}\nslope = {reg.coef_[0]:.3f}'
                plt.text(0.05, 0.95, stats_text, transform=plt.gca().transAxes, 
                        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
            
            plt.axhline(y=0, color='gray', linestyle='--')
            plt.axvline(x=0, color='gray', linestyle='--')
            plt.ylim(-350, 350)
            plt.xticks(fontsize=14)
            plt.yticks(fontsize=14)
            plt.xlabel("Visual Conflict (ms)", fontsize=16)
            plt.ylabel("PSE (ms)", fontsize=16)
            plt.title(f"Audio Noise Level: {audioNoiseLevel}", fontsize=16)
            plt.legend(title="Conflict Levels", fontsize=12, title_fontsize=12)

    plt.tight_layout()
    plt.show()
    
    return df_analysis, slopes_bootstrap
```

```python
# Additional Analysis: Effect of Visual Conflict on Sensitivity (Sigma)
def analyze_conflict_effects_on_sensitivity(best_fit, allBootedFits):
    """
    Analyze how visual conflict affects auditory sensitivity (sigma parameter)
    """
    print("\n" + "="*60)
    print("CONFLICT EFFECTS ON AUDITORY SENSITIVITY (σ)")
    print("="*60)
    
    sensitivity_data = []
    
    for j, audioNoiseLevel in enumerate(uniqueSensory):
        conflict_levels = []
        sigma_values = []
        
        for k, conflictLevel in enumerate(uniqueConflict):
            lambda_, mu, sigma = getParams(best_fit.x, conflictLevel, audioNoiseLevel, nLambda, nSigma)
            
            conflict_levels.append(conflictLevel * 1000)
            sigma_values.append(sigma)
            
            sensitivity_data.append({
                'noise_level': audioNoiseLevel,
                'conflict_ms': conflictLevel * 1000,
                'sigma': sigma
            })
        
        # Statistical analysis for sensitivity
        print(f"\nNoise Level {audioNoiseLevel}:")
        print("-" * 30)
        
        # Correlation between conflict and sigma
        pearson_r, pearson_p = pearsonr(conflict_levels, sigma_values)
        print(f"Conflict vs Sensitivity correlation: r = {pearson_r:.4f}, p = {pearson_p:.4f}")
        
        if pearson_p < 0.05:
            if pearson_r > 0:
                print("  → Higher conflict associated with LOWER sensitivity (higher σ)")
            else:
                print("  → Higher conflict associated with HIGHER sensitivity (lower σ)")
        else:
            print("  → No significant relationship between conflict and sensitivity")
    
    # Plot sensitivity analysis
    plt.figure(figsize=(12, 5))
    
    for j, audioNoiseLevel in enumerate(uniqueSensory):
        plt.subplot(1, 2, j + 1)
        
        subset = [d for d in sensitivity_data if d['noise_level'] == audioNoiseLevel]
        conflicts = [d['conflict_ms'] for d in subset]
        sigmas = [d['sigma'] for d in subset]
        
        colors = sns.color_palette("viridis", len(conflicts))
        
        plt.scatter(conflicts, sigmas, c=colors, s=100)
        
        # Add trend line if more than 2 points
        if len(conflicts) > 2:
            z = np.polyfit(conflicts, sigmas, 1)
            p = np.poly1d(z)
            x_trend = np.linspace(min(conflicts), max(conflicts), 100)
            plt.plot(x_trend, p(x_trend), "r--", alpha=0.8, linewidth=2)
            
            # Add correlation info
            r, p_val = pearsonr(conflicts, sigmas)
            plt.text(0.05, 0.95, f'r = {r:.3f}\np = {p_val:.3f}', 
                    transform=plt.gca().transAxes, verticalalignment='top',
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        plt.xlabel("Visual Conflict (ms)", fontsize=14)
        plt.ylabel("Sensitivity (σ)", fontsize=14)
        plt.title(f"Noise Level: {audioNoiseLevel}", fontsize=14)
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return pd.DataFrame(sensitivity_data)

# ANOVA for overall effects
def perform_anova_analysis(df_stats, df_sensitivity):
    """
    Perform ANOVA to test main effects and interactions
    """
    print("\n" + "="*60)
    print("ANOVA ANALYSIS")
    print("="*60)
    
    # Prepare data for ANOVA - remove any NaN or infinite values
    df_anova = df_stats.copy()
    df_anova = df_anova.dropna()  # Remove any NaN values
    df_anova = df_anova[np.isfinite(df_anova['pse_ms'])]  # Remove infinite values
    
    if len(df_anova) == 0:
        print("No valid data for ANOVA analysis after cleaning.")
        return None
    
    df_anova['noise_categorical'] = df_anova['noise_level'].astype('category')
    df_anova['conflict_categorical'] = pd.cut(df_anova['conflict_ms'], 
                                             bins=len(uniqueConflict), 
                                             labels=[f'C{i}' for i in range(len(uniqueConflict))])
    
    print("Two-way ANOVA: PSE ~ Noise Level * Conflict Level")
    print("-" * 50)
    
    try:
        # Fit ANOVA model
        formula = 'pse_ms ~ C(noise_categorical) + C(conflict_categorical) + C(noise_categorical):C(conflict_categorical)'
        model = sm.OLS.from_formula(formula, data=df_anova).fit()
        anova_table = anova_lm(model, typ=2)
        
        print(anova_table)
        
        # Effect sizes (eta squared)
        ss_total = anova_table['sum_sq'].sum()
        anova_table['eta_squared'] = anova_table['sum_sq'] / ss_total
        
        print(f"\nEffect Sizes (η²):")
        for index, row in anova_table.iterrows():
            if 'Residual' not in str(index):
                print(f"  {index}: η² = {row['eta_squared']:.4f}")
        
        # Post-hoc tests if significant main effects
        print(f"\nPost-hoc Analysis:")
        print("-" * 20)
        
        # Test each noise level separately
        for noise in uniqueSensory:
            subset = df_anova[df_anova['noise_level'] == noise]
            if len(subset) > 2:
                # One-way ANOVA for conflict effect at this noise level
                groups = [subset[subset['conflict_categorical'] == cat]['pse_ms'].values 
                         for cat in subset['conflict_categorical'].unique() if not pd.isna(cat)]
                
                if len(groups) > 1 and all(len(g) > 0 for g in groups):
                    f_stat, p_val = stats.f_oneway(*groups)
                    print(f"  Noise {noise} - Conflict effect: F = {f_stat:.4f}, p = {p_val:.4f}")
        
        return anova_table
    
    except Exception as e:
        print(f"ANOVA analysis failed: {e}")
        print("Performing simplified analysis instead...")
        
        # Simplified analysis using t-tests
        for noise in uniqueSensory:
            subset = df_anova[df_anova['noise_level'] == noise]
            if len(subset) > 2:
                # Test if PSE differs significantly from 0
                t_stat, p_val = stats.ttest_1samp(subset['pse_ms'], 0)
                print(f"  Noise {noise} - PSE vs 0: t = {t_stat:.4f}, p = {p_val:.4f}")
        
        return None

# Summary statistics table
def create_summary_table(df_stats):
    """
    Create a comprehensive summary statistics table
    """
    print("\n" + "="*60)
    print("SUMMARY STATISTICS TABLE")
    print("="*60)
    
    # Clean the data first
    df_clean = df_stats.dropna()
    
    if len(df_clean) == 0:
        print("No valid data for summary table.")
        return None
    
    summary = df_clean.groupby(['noise_level', 'conflict_ms']).agg({
        'pse_ms': ['mean', 'std', 'count'],
        'pse_ci_lower': 'mean',
        'pse_ci_upper': 'mean'
    }).round(3)
    
    summary.columns = ['PSE_mean', 'PSE_std', 'N', 'CI_lower', 'CI_upper']
    summary['CI_width'] = summary['CI_upper'] - summary['CI_lower']
    
    print(summary)
    
    # Export to CSV for further analysis
    summary.to_csv('pse_summary_statistics.csv')
    print(f"\nSummary table saved to: pse_summary_statistics.csv")
    
    return summary
```
